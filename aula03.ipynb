```python
# chrome > chromedriver
# firefox > geckodriver
# enviar arquivo do webdrive para pasta do python
# chamar informações da bibilioteca de forma prática
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By


# Cria o navegador
nav = webdriver.Chrome()

# entrar no google e pesquisar a cotação do dolar
nav.get("https://www.google.com.br/")
nav.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("Cotação do dolar", Keys.ENTER)

# pegar a cotação do dólar
cotDolar = nav.find_element(By.XPATH, '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute("data-value")
print(f'1 U$ = {cotDolar:.4} R$')

#=========================================

# pesquisar no google a contação do euro
nav.get("https://www.google.com.br/")
nav.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("Cotação do euro", Keys.ENTER)


# pegar a cotação do euro
cotEuro = nav.find_element(By.XPATH, '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute("data-value")
print(f'1 € = {cotEuro:.4} R$')

# =======================================

# pegar a cotação do ouro
nav.get("https://www.melhorcambio.com/ouro-hoje")
cotOuro = nav.find_element(By.XPATH, '//*[@id="comercial"]').get_attribute('value')
cotOuro = cotOuro.replace(',', '.')

print(f'1 G de ouro = {cotOuro:.6} R$')


nav.quit() # encerrar navegador da função
```

Importando a base de dados

```python
# importar a base de dados
import pandas as pd

tabela = pd.read_excel("Produtos.xlsx")
display(tabela)
```

Atualizando os preços e o cáculo do preço final

```python
# atualizar cotação do dolar
locDolar = tabela["Moeda"] == "Dólar"
tabela.loc[locDolar, "Cotação"] = float(cotDolar)#localiza as linha e colunas e prenche com um valor especifico

# atualizar cotação do euro
locDolar = tabela["Moeda"] == "Euro"
tabela.loc[locDolar, "Cotação"] = float(cotEuro)

# atualizar cotação do ouro
locDolar = tabela["Moeda"] == "Ouro"
tabela.loc[locDolar, "Cotação"] = float(cotOuro)

# atualizando valores

# atualizar preço de compra  (preço original * cotação)
tabela["Preço de Compra"] = tabela["Preço Original"] * tabela["Cotação"]
# atualizar preço de venda  (preço de compra * margem)
tabela["Preço de Venda"] = tabela["Preço de Compra"] * tabela["Margem"]

# Exibir dados atualizados
display(tabela)
```

Exportar a nova base de preço atualizada

```python
# Exportar essa base de dados para ter o resultado atualizado

tabela.to_excel("ProdutosNovo.xlsx", index=False)
```
